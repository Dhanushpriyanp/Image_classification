```python
import time
from ultralytics import YOLO
import ultralytics
import cv2, os
import numpy as np
```

```python
path = "Images"
```

```python
model_bead = YOLO("Models/v18.pt")
model_stone = YOLO("Models/v13v9.pt")
model_tassel = YOLO("Models/v10.pt")
```

```python
results_bead = model_bead.predict(source=path, imgsz=640, conf=0.5)
```

```python
results_stone = model_stone.predict(source=path, imgsz=640, conf=0.5)
# results_stone =results_bead
```

```python
# results_tassel = model_tassel.predict(source=path, imgsz=640, conf=0.5)
results_tassel = results_bead
```

```python
image_c = 0
bead_c = 0
stone_c = 0
tassel_c = 0
```

```python
import os
import cv2

# Create a directory to store the predicted images
directory = "predicted_images1"
os.makedirs(directory, exist_ok=True)

for result in range(len(results_stone)):
    # print(results_stone[i],results_bead[i],results_tassel[i])
    boxes_stone = results_stone[result].boxes
    boxes_bead = results_bead[result].boxes
    boxes_tassel = results_tassel[result].boxes
    # print(boxes)

    image = cv2.imread(results_stone[result].path)
    image_c += 1
    # Define colors for different classes (adjust as needed)
    stone_color = (0, 255, 0)  # Green
    bead_color = (0, 0, 255)
    tassel_color = (255, 0, 0)
    # bead_confidences = boxes.conf[boxes.cls == 1]
    # stone_beads = bead_confidences[bead_confidences >= 0.3]
    c=0
    if len(boxes_stone.cls[(boxes_stone.cls == 1)]) > 0:
        c+=1
        stone_indices = boxes_stone.cls == 1
        stone_xys = boxes_stone.xyxy[stone_indices]

        conf = boxes_stone.conf[stone_indices]
        stone_c += 1
        for s1 in results_stone[result].verbose().split(","):
            if "stone" in s1:
                print(s1,end=" ")
        # Loop through high-confidence stone boxes
        for i in range(len(stone_xys)):

            x1, y1, x2, y2 = stone_xys[i].cpu().numpy().astype(int)
            cv2.rectangle(image, (x1, y1), (x2, y2), stone_color, 2)
            cv2.putText(
                image,
                f"Stones {conf[i]:.2f}",
                (int(x1), int(y1 - 5)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                stone_color,
                2,
            )
    if len(boxes_bead.cls[(boxes_bead.cls == 0)]) > 0:
        c+=1
        bead_indices = boxes_bead.cls == 0
        bead_xys = boxes_bead.xyxy[bead_indices]

        conf = boxes_bead.conf[bead_indices]
        bead_c += 1
        for b1 in results_bead[result].verbose().split(","):
            if "bead" in b1:
                print(b1,end=" ")
        # Loop through high-confidence bead boxes
        for i in range(len(bead_xys)):

            x1, y1, x2, y2 = bead_xys[i].cpu().numpy().astype(int)
            cv2.rectangle(image, (x1, y1), (x2, y2), bead_color, 2)
            cv2.putText(
                image,
                f"Beads {conf[i]:.2f}",
                (int(x1), int(y1 - 5)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                bead_color,
                2,
            )

    if len(boxes_tassel.cls[(boxes_tassel.cls == 2)]) > 0:
        c+=1
        tassel_indices = boxes_tassel.cls == 2
        tassel_xys = boxes_tassel.xyxy[tassel_indices]

        conf = boxes_tassel.conf[tassel_indices]
        tassel_c += 1
        for t1 in results_tassel[result].verbose().split(","):
            if "tassel" in t1:
                print(t1,end=" ")
        # Loop through high-confidence tassel boxes
        for i in range(len(tassel_xys)):

            x1, y1, x2, y2 = tassel_xys[i].cpu().numpy().astype(int)
            cv2.rectangle(image, (x1, y1), (x2, y2), tassel_color, 2)
            cv2.putText(
                image,
                f"Tassels {conf[i]:.2f}",
                (int(x1), int(y1 - 5)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                tassel_color,
                2,
            )
    if c>0:
        print(results_stone[result].path)

    image_name = os.path.basename(results_stone[result].path)
    image_path = os.path.join(directory, image_name)
    cv2.imwrite(image_path, image)

    # Print message indicating image is stored
    print(f"Image {image_name} is stored in the directory: {directory}")

print(image_c,bead_c,stone_c,tassel_c)
print(bead_c/image_c)#0.85
print(stone_c/image_c)#0.85
print(tassel_c/image_c)#0.96
```
